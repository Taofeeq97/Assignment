```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TCC.csv')
```

```python
df.count()
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
X= df.drop(['customerID', 'Churn'], axis=1)
y= df['Churn']
```

```python
X.info()
```

```python
x = pd.get_dummies( columns=['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'], drop_first=True, data=X, dtype=int)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
len(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_test_sc
```

KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_sc, y_train)
```

```python
y_pred = knn.predict(x_test_sc)
```

```python
from sklearn.metrics import accuracy_score
accuracy = (accuracy_score(y_test, y_pred) * 100)
accuracy
```

```python
from sklearn.metrics import recall_score, precision_score
knn_precision = (precision_score(y_test, y_pred, pos_label='Yes') * 100)
print(knn_precision)
knn_recall = (recall_score(y_test, y_pred, pos_label='Yes') * 100)
print(knn_recall)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train_sc, y_train)
```

```python
yd_pred = dt.predict(x_test_sc)

accuracy = (accuracy_score(y_test, yd_pred) * 100)
accuracy
```

```python
from sklearn.metrics import recall_score, precision_score
dt_precision = (precision_score(y_test, yd_pred, pos_label='Yes') * 100)
print(dt_precision)
dt_recall = (recall_score(y_test, yd_pred, pos_label='Yes') * 100)
print(dt_recall)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000)
rf.fit(x_train_sc, y_train)
```

```python
yr_pred = rf.predict(x_test_sc)

accuracy = (accuracy_score(y_test, yr_pred) * 100)
accuracy
```

```python
rf_precision = (precision_score(y_test, yr_pred, pos_label='Yes') * 100)
print(rf_precision)
rf_recall = (recall_score(y_test, yr_pred, pos_label='Yes') * 100)
print(rf_recall)
```

```python
from sklearn.metrics import classification_report
print('Random Forest')
print(classification_report(y_test, yr_pred))
```

```python
print('KNN')
print(classification_report(y_test, y_pred))
```

```python
print('Decision Tree')
print(classification_report(y_test, yd_pred))
```

Bayes Naives

```python
from sklearn.naive_bayes import BernoulliNB
nb = BernoulliNB()
nb.fit(x_train_sc, y_train)
```

```python
ynb_pred = nb.predict(x_test_sc)

accuracy = (accuracy_score(y_test, ynb_pred) * 100)
accuracy
```

```python
print('Naive Bayes')
print(classification_report(y_test, ynb_pred))
```

SVM

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train_sc, y_train)
```

```python
ysvm_pred = svm.predict(x_test_sc)

accuracy = (accuracy_score(y_test, ysvm_pred) * 100)
accuracy
```

```python
print('SVM')
print(classification_report(y_test, ysvm_pred))
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_sc, y_train)
```

```python
ylr_pred = lr.predict(x_test_sc)

accuracy = (accuracy_score(y_test, ylr_pred) * 100)
accuracy
```

```python
print('logistic regression')
print(classification_report(y_test, ylr_pred))
```
